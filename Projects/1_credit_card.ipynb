 # Following is the Data Dictionary for Credit Card dataset :-

* CUST_ID : Identification of Credit Card holder (Categorical)
* BALANCE : Balance amount left in their account to make purchases (
* BALANCE_FREQUENCY : How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not frequently updated)
* PURCHASES : Amount of purchases made from account
* ONEOFF_PURCHASES : Maximum purchase amount done in one-go
* INSTALLMENTS_PURCHASES : Amount of purchase done in installment
* CASH_ADVANCE : Cash in advance given by the user
* PURCHASES_FREQUENCY : How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)
* ONEOFFPURCHASESFREQUENCY : How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased)
* PURCHASESINSTALLMENTSFREQUENCY : How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done)
* CASHADVANCEFREQUENCY : How frequently the cash in advance being paid
* CASHADVANCETRX : Number of Transactions made with "Cash in Advanced"
* PURCHASES_TRX : Numbe of purchase transactions made
* CREDIT_LIMIT : Limit of Credit Card for user
* PAYMENTS : Amount of Payment done by user
* MINIMUM_PAYMENTS : Minimum amount of payments made by user
* PRCFULLPAYMENT : Percent of full payment paid by user
* TENURE : Tenure of credit card service for user

```python
# checking for outliers in credit limit
cc_df['CREDIT_LIMIT'].plot(kind = 'box')
```

* OBSERVATION : THIS TOO HAVE OULIERS SO MEDIAN HERE IS THE SOLUTION

```python
cc_df['MINIMUM_PAYMENTS'] = cc_df['MINIMUM_PAYMENTS'].fillna(cc_df['MINIMUM_PAYMENTS'].median())
```

# FEATURE ENGINEERING


```python
cc_df.columns
```

```python
# --- Feature engineering block (insert after cleaning, before scaling) ---

fe = cc_df.copy()

# Utilization and payment behavior
fe['credit_utilization']     = fe['BALANCE'] / (fe['CREDIT_LIMIT'] + 1)
fe['payment_ratio']          = fe['PAYMENTS'] / (fe['BALANCE'] + 1)
fe['purchase_to_limit']      = fe['PURCHASES'] / (fe['CREDIT_LIMIT'] + 1)

# Spend composition
fe['cash_advance_ratio']     = fe['CASH_ADVANCE'] / (fe['PURCHASES'] + fe['CASH_ADVANCE'] + 1)
fe['installment_ratio']      = fe['INSTALLMENTS_PURCHASES'] / (fe['PURCHASES'] + 1)
fe['oneoff_ratio']           = fe['ONEOFF_PURCHASES'] / (fe['PURCHASES'] + 1)

# Per-transaction averages
fe['avg_purchase_per_txn']   = fe['PURCHASES'] / (fe['PURCHASES_TRX'] + 1)
fe['avg_cash_adv_per_txn']   = fe['CASH_ADVANCE'] / (fe['CASH_ADVANCE_TRX'] + 1)

# Activity level
fe['total_transactions']     = fe['PURCHASES_TRX'] + fe['CASH_ADVANCE_TRX']
fe['spend_to_payment']       = (fe['PURCHASES'] + fe['CASH_ADVANCE']) / (fe['PAYMENTS'] + 1)

print("New features added:", 10)
print(fe.shape)  # (8950, 27)



from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(fe)
X_scaled = pd.DataFrame(X_scaled, columns=fe.columns)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertias = []
silhouettes = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, km.labels_))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Elbow plot
axes[0].plot(K_range, inertias, 'o-', color='steelblue')
axes[0].set_xlabel('Number of clusters (K)')
axes[0].set_ylabel('Inertia')
axes[0].set_title('Elbow method')

# Silhouette plot
axes[1].plot(K_range, silhouettes, 'o-', color='coral')
axes[1].set_xlabel('Number of clusters (K)')
axes[1].set_ylabel('Silhouette score')
axes[1].set_title('Silhouette scores')

plt.tight_layout()
plt.show()

print("Best K by silhouette:", K_range[silhouettes.index(max(silhouettes))])
```

```python
BEST_K = 3 
km = KMeans(n_clusters=BEST_K, random_state=42, n_init=10)
km.fit(X_scaled)

fe['Cluster'] = km.labels_
print(fe['Cluster'].value_counts().sort_index())


centers = scaler.inverse_transform(km.cluster_centers_)
centers_df = pd.DataFrame(centers, columns=fe.columns[:-1])
print(centers_df.round(2))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled.drop('Cluster', axis=1, errors='ignore'))

print(f"Variance explained: {pca.explained_variance_ratio_.sum()*100:.1f}%")

plt.figure(figsize=(9, 6))
colors = ['#378ADD', '#1D9E75', '#D85A30', '#7F77DD']

for cluster_id in range(BEST_K):
    mask = fe['Cluster'] == cluster_id
    plt.scatter(
        X_pca[mask, 0], X_pca[mask, 1],
        c=colors[cluster_id], label=f'Cluster {cluster_id}',
        alpha=0.5, s=15
    )

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.title('K-Means clusters in PCA space')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Mean of each feature per cluster
profile = fe.groupby('Cluster').mean().round(2)
print(profile.T)

# Heatmap of normalized cluster profiles
profile_norm = (profile - profile.min()) / (profile.max() - profile.min())

plt.figure(figsize=(10, 5))
sns.heatmap(
    profile_norm.T,
    annot=True, fmt='.2f',
    cmap='YlOrBr',
    linewidths=0.5,
    cbar_kws={'label': 'Normalized value (0–1)'}
)
plt.title('Cluster profiles — normalized feature means')
plt.tight_layout()
plt.show()

# Typical segments you might find:
# Cluster 0 — "Revolvers":   high balance, low payments
# Cluster 1 — "Transactors": high purchases, low cash advance
# Cluster 2 — "Cash users":  high cash advance, low purchases
# Cluster 3 — "Inactives":   low activity across all features
```
